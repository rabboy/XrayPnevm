```python
#импортирую модуль os для доступа к файлам
import os
#импортирую модуль numpy и подписываю его как np для создания массивов и взаимодействия с ними
import numpy as np
#импортирую модуль tensorflow и подписываю его как tf для создания модели 
import tensorflow as tf
#импортирую последовательную модель Sequential
from tensorflow.keras.models import Sequential
#импортирую слои Conv2D, MaxPooling2D, Dense, Flatten, Dropout
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout
from tensorflow.keras import utils
from tensorflow.keras.preprocessing import image
#импортирую функцию подгрузки фото
from tensorflow.keras.preprocessing import image_dataset_from_directory
#импортирую pyplot как plt для отрисовки графиков и показа фотографий с подписями классов
import matplotlib.pyplot as plt
#чтобы строить графики прямо внутри ноутбука
%matplotlib inline 
```

```python
#указываю размер батча и размер изображения
batch_size=256
image_size=(100, 100)
```

```python
#подгружаю папку с тренировочными фотографиями
train_dataset = image_dataset_from_directory('arc/chest_xray/train',
                                             subset='training',
                                             seed=42,
                                             validation_split=0.1,
                                             batch_size=batch_size,
                                             image_size=image_size)
```

```python
#создаю переменную, которая содержит в себе имена классов фото и вывожу их
class_names = train_dataset.class_names
class_names
```

```python
#задаю размер фигуры и рисую 9 фотографий с подписями классов
plt.figure(figsize=(8, 8))
for images, labels in train_dataset.take(1):
  for i in range(9):
    ax = plt.subplot(3, 3, i + 1)
    plt.imshow(images[i].numpy().astype("uint8"))
    plt.title(class_names[labels[i]])
    plt.axis("off")
```

```python
#подгружаю тестовый датасет
test_dataset = image_dataset_from_directory('arc/chest_xray/test',
                                             batch_size=batch_size,
                                             image_size=image_size)
```

```python
#вывожу названия классов в тестовом датасете
test_dataset.class_names
```

```python
#подгружаю валидирующий датасет
validation_dataset = image_dataset_from_directory('arc/chest_xray/val',
                                             subset='validation',
                                             seed=42,
                                             validation_split=0.1,
                                             batch_size=batch_size,
                                             image_size=image_size)
```

```python
#используем автоматические подобранные параметры для лучшей работы нейросети
AUTOTUNE = tf.data.experimental.AUTOTUNE

train_dataset = train_dataset.prefetch(buffer_size=AUTOTUNE)
validation_dataset = validation_dataset.prefetch(buffer_size=AUTOTUNE)
test_dataset = test_dataset.prefetch(buffer_size=AUTOTUNE)
```

```python
# Создаем последовательную модель и дальше добавляем в неё слои
model = Sequential()
# Сверточный слой Conv2D с 16 фильтрами и размером ядра 5*5 каждый, функция активации relu и выходная карта признаков на каждом канале той же размерности, что и исходное изображение, т.е 100*100
model.add(Conv2D(16, (5, 5), padding='same', 
                 input_shape=(100, 100, 3), activation='relu'))
# Слой подвыборки MaxPooling2d, который преобразовывает множество значений  в единое значение — максимальное среди выборки. Указан размер пула(множества значений подвыборки) - 2*2
model.add(MaxPooling2D(pool_size=(2, 2)))
# Сверточный слой Conv2D с 32 фильтрами и размером ядра 5*5 каждый, функция активации relu и выходная карта признаков на каждом канале той же размерности, что и исходное изображение
model.add(Conv2D(32, (5, 5), activation='relu', padding='same'))
# Слой подвыборки
model.add(MaxPooling2D(pool_size=(2, 2)))
# Сверточный слой Conv2D с 64 фильтрами и размером ядра 5*5 каждый, функция активации relu и выходная карта признаков на каждом канале той же размерности, что и исходное изображение
model.add(Conv2D(64, (5, 5), activation='relu', padding='same'))
# Слой подвыборки
model.add(MaxPooling2D(pool_size=(2, 2)))
# Сверточный слой Conv2D с 128 фильтрами и размером ядра 5*5 каждый, функция активации relu и выходная карта признаков на каждом канале той же размерности, что и исходное изображение
model.add(Conv2D(128, (5, 5), activation='relu', padding='same'))
# Слой подвыборки
model.add(MaxPooling2D(pool_size=(2, 2)))
# Полносвязная часть нейронной сети для классификации
# слой, который преобразовывает многомерный входной тензор в одномерный тензор 
model.add(Flatten())
# слой, который объединяет нейроны из предыдущего и следующего слоя, функция активации relu
model.add(Dense(1024, activation='relu'))
# слой для "борьбы" с переобучением
model.add(Dropout(0.2))
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.2))
# Выходной слой, 2 нейрона по количеству классов
model.add(Dense(2, activation='softmax'))
```

```python
#собираю модель с указанием функции потерь, оптимизатором и метрикой
model.compile(loss='sparse_categorical_crossentropy',
              optimizer="adam",
              metrics=['accuracy'])
```

```python
#обучаю модель за 4 эпохи с указанием обучающего и валидирующего датасетов
history = model.fit(train_dataset, 
                    validation_data=validation_dataset,
                    epochs=4,
                    verbose=2)
```

```python
# Оцениваем качество обучения модели на тестовых данных
scores = model.evaluate(test_dataset, verbose=1)
```

```python
#вывожу долю верных ответов на тестовых данных
print("Доля верных ответов на тестовых данных, в процентах:", round(scores[1] * 100, 4))
```

```python
#черчу график, который показывает как обучилась модель
plt.plot(history.history['val_accuracy'], 
         label='Доля верных ответов на обучающем наборе')
plt.plot(history.history['accuracy'], 
         label='Доля верных ответов на проверочном наборе')
plt.xlabel('Эпоха обучения')
plt.ylabel('Доля верных ответов')
plt.legend()
plt.show()
```
